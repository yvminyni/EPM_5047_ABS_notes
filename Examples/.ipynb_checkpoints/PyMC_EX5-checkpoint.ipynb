## Ex5. Hierarchical model... Heart Attack

Example 5.

The New York State (NYS) Department of Health collects and releases data on mortality after Acute Myocardial Infarction (AMI), commonly known as a heart attack. Their 2015 report was the initial public data release by the NYS Department of Health on risk-adjusted mortality outcomes for AMI patients at hospitals across the state. We focus on 13 hospitals in Manhattan, New York City, with the goal of learning about the percentages of resulted deaths from heart attack for hospitals in this sample$.^{[1],[2]}$ 

The data records 13 hospitals in New York City - Manhattan in 2015. It contains the number of heart attack cases, the corresponding number of resulted deaths, and their computed percentage of resulted deaths.


```python
import os
#file location (on desktop)
filename =os.environ["HOMEPATH"] + "\Desktop\DeathHeartAttackManhattan.csv" 

df = pd.read_csv(filename)
df['Hospital']=df['Hospital'].astype('category')
df
```

Example 5.

Heart Attack Deaths - New York City

Likelihood:

\begin{align*}
      Y_i \mid n_i, p_i &\sim \text{Binomial}(n_i, p_i) \\ 
\end{align*} 

Prior for $p_i$, for $i=1,...,13$:
\begin{align*}
      p_i \mid \alpha, \beta  &\sim \text{Beta}(\alpha  , \beta)\\
\end{align*}

Prior for $p_i$, hyperprior:
\begin{align*}
 \alpha,\beta, &\sim \pi(\alpha  , \beta)
\end{align*}
\begin{align*}
      \alpha &\sim \text{Gamma}( 4, 0.5) \\
      \beta &\sim \text{Gamma}( 4, 0.5) 
\end{align*}

- $Y_i$: The number of resulted deaths from heart attack.
- $n_i$: The number of heart attack cases
- $p_i$: The death rate for hospital $i$

The columns we need are `Cases`, which is the number of patients treated at each hospital, and `Deaths`, which is the number of those patients who died.

```python
#number of case
data_ns = df['Cases'].values #cases (n_i) number of patients treated at each hospital
#died 
data_ds = df['Deaths'].values #successes (y_i) number of those patients who died

#Index: Array, dtype=int8 0,1,2,3
df_idx = df.Hospital.cat.codes.values 
#Label: Array, object (category)['h1', 'h10', 'h11',...]
df_labels =  df.Hospital.cat.categories 
```

```python
coords = {"Hospital": df_labels,        # Label/Name for the hospital 
          "Hosp_idx":df_idx,            # Index for the hospital
         "obs_id": df_labels[df_idx]} # Index for the dataframe
```

```python
with pm.Model(coords=coords) as m_heartattack:
    ##hyper-prior
    alpha = pm.Gamma('alpha', alpha=4, beta=0.5)
    beta = pm.Gamma('beta', alpha=4, beta=0.5)
    #prior
    ###p = pm.Beta('pi', alpha, beta, shape=13)
    pi = pm.Beta('pi', alpha, beta, dims="Hospital") 
    #likelihood
    yi = pm.Binomial('yi', n=data_ns, p=pi[df_idx], observed=data_ds, dims="obs_id")
```

```python
pm.model_to_graphviz(model=m_heartattack)
```

```python
gv = pm.model_graph.model_to_graphviz(model=m_heartattack)
gv.format = 'png'
gv.render(filename='ex_5_heart')
```

```python
with m_heartattack: 
    idata_heart = pm.sample(1000, random_seed=101)
az.summary(idata_heart, hdi_prob=0.95).round(2)
```

```python
az.plot_posterior(idata_heart, var_names=["~alpha","~beta"], hdi_prob=0.95,  round_to=2)
plt.subplots_adjust(hspace=0.6)
plt.show()
```

```python
az.plot_trace(idata_heart,compact=True )
plt.tight_layout()
```

```python
summary = az.summary(idata_heart, var_names=["pi"], kind="stats") #p_i: death rate
posterior_means = summary["mean"] 
print(posterior_means)
```

```python
df_combined = pd.concat([df,  posterior_means.iloc[df_idx].reset_index(drop=True)*100], axis=1).rename(columns={"mean": "$p_i$ posterior mean(%)"})
df_combined 
```
